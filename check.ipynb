```python
import pandas as pd
from sqlalchemy import create_engine

# Подключение к PostgreSQL
db_user = "postgres"
db_password = ""
db_host = "localhost"
db_port = "5432"
db_name = "instacart_analysis"

engine = create_engine(f"postgresql+psycopg2://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}")
```

```python
aisles = ["aisles"]

for ais in aisles:
    df = pd.read_sql(f"SELECT * FROM {ais} LIMIT 1000", engine)
    print(f"\n--- {ais} ---")
    print(df.info())      
    print(df.describe())  
    print(df.head())      
```
